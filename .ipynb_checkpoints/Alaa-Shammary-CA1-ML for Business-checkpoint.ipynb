The forward fill method is used because customer ID's are grouped together so its more accurate to use the forward method for accuracy of data.

```python
data_df['Customer ID'].fillna(method='ffill',inplace=True)
```

Unknown is used to fill missing values because every product is unique and is text based so the use of the mode or forward fill would create some bias in the dataset analysis.

```python
data_df['Description'].fillna('Unknown',inplace=True)
```

```python
data_df.isnull().sum()
```

drop categorical data

```python
columns_df = data_df[['Quantity', 'Price', 'Customer ID']]
```

standard scaler

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(columns_df)
```

```python
columns_df.head()
```

```python
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
k_clusters = kmeans.fit_predict(scaled)
```

```python
wcss = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(columns_df)
    wcss.append(kmeans.inertia_)
print(wcss)
```

```python
plt.plot(range(2,10), wcss);
```

```python
columns_df['Cluster'] = kmeans.labels_
```

```python
def plotFeatures(col1, col2):
    plt.scatter(
        columns_df[kmeans.labels_ == 0].iloc[:, col1],
        columns_df[kmeans.labels_ == 0].iloc[:, col2],
        s=50,
        c='red',
        label='Cluster 1'
    )
    plt.scatter(
        columns_df[kmeans.labels_ == 1].iloc[:, col1],
        columns_df[kmeans.labels_ == 1].iloc[:, col2],
        s=50,
        c='blue',
        label='Cluster 2'
    )
    plt.scatter(
        columns_df[kmeans.labels_ == 2].iloc[:, col1],
        columns_df[kmeans.labels_ == 2].iloc[:, col2],
        s=50,
        c='green',
        label='Cluster 3'
    )
    plt.scatter(
        kmeans.cluster_centers_[:, col1],
        kmeans.cluster_centers_[:, col2],
        s=100,
        c='yellow',
        label='Centroids'
    )
    plt.xlabel(columns_df.columns[col1])
    plt.ylabel(columns_df.columns[col2])
    plt.legend()
    plt.title('Clustering based on ' + columns_df.columns[col1] + ' and ' + columns_df.columns[col2])
    plt.show()

plotFeatures(0, 1)  
```

```python
plotFeatures(0,2)
```

```python
plotFeatures(1,2)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn_extra.cluster import KMedoids
```

```python
df = data_df[['Quantity','Price']]
```

```python
df.head()
```

```python
standard = StandardScaler()
```

```python
x = standard.fit_transform(df)
```

```python
coumns = df.columns
```

```python
coumns
```

```python
df = pd.DataFrame(standard.fit_transform(df),columns=coumns)
```

```python
df.head()
```

K-elbow method

```python
wcss = []
for k in range (2,10):
    kmeans = KMeans(n_clusters = k,random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(2,10),wcss);    
```

Silhouette Score

```python
wcss = []
for k in range (2,10):
    kmeans = KMeans(n_clusters = k,random_state=42)
    y_kmeans = kmeans.fit_predict(df)
    wcss.append(silhouette_score(df,y_kmeans))
plt.plot(range(2,10),wcss);    
```
